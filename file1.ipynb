```python
import pandas as pd
```

```python
df=pd.read_csv('Messy_Employment_India_Dataset.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['Years of Experience'],inplace=True)
df.isnull().sum()
```

```python
categorical_cols=['Status','Age Group','Education','Industry','Location','AI Risk']
df[categorical_cols]=df[categorical_cols].fillna('Unknown')
df.isnull().sum()
```

```python
df['Monthly Salary (INR)']=df['Monthly Salary (INR)'].fillna(df['Monthly Salary (INR)'].median())
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
df['date_recorded']=pd.to_datetime(df['date_recorded'])
df['date_recorded']
```

```python
df['month_recorded']=df['date_recorded'].dt.month_name()
df['year_recorded']=df['date_recorded'].dt.year
df.head()
```

```python
df.drop(columns=['date_recorded'],inplace=True)
```

```python
df.head()
```

```python
df.to_csv('Clean_Employment_India_Dataset.csv',index=False)
```
